## Setup

Run this cell first. It imports the checker used by the function cells.

````python
import sys
from pathlib import Path

sys.dont_write_bytecode = True

def find_project_root():
    for path in (Path.cwd(), *Path.cwd().parents):
        if (path / "course_utils").is_dir() and (path / "pyproject.toml").is_file():
            return path
    raise RuntimeError(
        "Could not find the project root. Open the yisen-coding folder in VS Code, "
        "or start Jupyter with `uv run jupyter lab` from the yisen-coding folder."
    )

print("Initializing the project paths.")

PROJECT_ROOT = find_project_root()
HOMEWORK_DIR = PROJECT_ROOT / "homework" / "hw01"

print(f"Found project root: {PROJECT_ROOT}")

if str(PROJECT_ROOT) not in sys.path:
    sys.path.insert(0, str(PROJECT_ROOT))

print(f"Successfully added the root to path.")


from course_utils.notebook_check import check, trace

CASES = "homework.hw01.tests.cases"
TRACES = "homework.hw01.tests.traces"

print(f"Successfully loaded the homework checker.")
print(f"Happy Coding!")
````

## Problem 1

Write a function named `greet` that takes a name and returns a greeting.

Example: `greet("Ada")` should return `"Hello, Ada!"`.

````python
@check("greet", CASES)
def greet(name):
    return f"Hello, {name}!"
````

## Problem 2

Write a function named `add` that returns the sum of two numbers.

````python
@check("add", CASES)
def add(a, b):
    return a + b
````

## Problem 3

Write a function named `is_even` that returns `True` when a number is even and `False` when it is odd.

````python
@check("is_even", CASES)
def is_even(number):
    return number % 2 == 0
````
