## 1.1 Types

First, let's learn about Python types.

Every Python value has a **type**. A type is the category Python uses to understand what a value is and what we can do with it.

For example, `123`, `67.69`, `True`, and `"hello"` are all values, but they do not all have the same type.

In this section, use this habit:
1. Predict the type.
2. Run `type(...)`.
3. Read Python's answer.

Before running the cell, predict the type of each value.

```python
print(type(123))
print(type(-3))
print(type(10086))
```

`123`, `-3`, and `10086` all have type `int`, which is short for **integer**.

An `int` is a whole number with no decimal point.

Python displays the answer as `<class 'int'>`. For now, just read that as `int`.

**Notes:**
- Integers can be positive, negative, or zero.
- Integers can have many digits, like `78238723874327827348724783`.
- `42` is an `int`, but `42.` is not. That decimal point matters.

Again, predict first, then run the cell.

```python
print(type(67.69))
print(type(42.))
print(type(4.1e+48))
```

`67.69`, `42.`, and `4.1e+48` all have type `float`.

A `float`, also called a **floating-point number**, is usually a number with a decimal point.

**Notes:**
- Any number that contains a decimal point is a `float`, so `42.` is a `float` and not an `int`.
- `42` and `42.` look similar to us, but Python sees different types.
- Python uses scientific notation for very large or very small floats. For example, `4.1e+48` means `4.1 * 10^48`.

Not every value is a number. Python also has Booleans, which represent true-or-false values.

```python
print(type(True))
print(type(False))
print(True)
print(False)
```

`True` and `False` both have type `bool`, short for **Boolean**.

**Notes:**
- `True` and `False` are the only Boolean values.
- The capital letters matter: use `True` and `False`, not `true` and `false`.
- We will use Booleans later when code needs to answer yes-or-no questions.

Next, let's look at strings. Strings are used for text.

```python
print(type('abc'))
print(type("Yisen"))
print(type(""))
```

`'abc'`, `"Yisen"`, and `""` all have type `str`, which is short for **string**.

Strings are used to represent text, or any sequence of characters.

**Notes:**
- A string usually uses single quotes, like `'abc'` or `'This is a string!'`.
- You can also use double quotes, so `"a"` and `"ab"` are also strings.
- The empty string, `""`, is a string containing no characters. Its length is `0`, but it is still a string.
- Anything inside quotes is a string, even if it looks like a number or a Boolean.

This is an important difference. Compare the values with quotes and without quotes:

```python
print("Without quotes:")
print(type(123))
print(type(3.14))
print(type(True))

print("With quotes:")
print(type("123"))
print(type("3.14"))
print(type("True"))
```

## 1.2 Errors

A quick note about errors. There are 3 common kinds of programming errors:

1. **Syntax Errors**
2. **Runtime Errors**
3. **Logical Errors**

### 1. Syntax Errors

Syntax errors happen when your code does not follow Python's grammar rules. Python notices the problem before the code starts running.

For example, try running this cell:

```python
# print(123
```

The code is missing the right parenthesis `)`, so Python gives a `SyntaxError` and tells us which line has the problem.

**Other syntax error examples:**
- Missing a quote, like `print("hello)`
- Missing a parenthesis, like `print(type(42)`
- Putting two values beside each other with no operator, like `print(4 5)`

```python
# print("Hello)
```

```python
# print(4 5)
```

### 2. Runtime Errors

Runtime errors happen when your code starts running, but then something goes wrong and the code stops.

For example:

```python
print(42)
print(1 / 0)
print("This line will not run")
```

The code runs and prints `42`, but then it crashes on line 2 because division by zero is not allowed. The last line does not run because Python stops at the error.

Here is another runtime error:

```python
# print("The answer is " + 42)
```

This crashes because Python cannot add a `str` and an `int` with `+`. The code is valid Python grammar, but it fails while running.

### 3. Logical Errors

Logical errors happen when your code runs without crashing, but it does not do what you intended. These are often the hardest errors to find because Python does not know what you meant.

For example, suppose we want the average of `80` and `100`:

```python
print("The average of 80 and 100 should be 90.")
print("My answer is:")
print(80 + 100 / 2)
```

This code runs, but it prints the wrong answer. Python does division before addition, so it calculates `100 / 2` first, then adds `80`.

The intended expression needs parentheses:

```python
print((80 + 100) / 2)
```

Here is another logical error that is about types:

```python
print("I want the type of the number 123.")
print(type("123"))
```

This code runs, but it checks the type of the string `"123"`, not the integer `123`. The quotes changed the value into a `str`.

**Remember:**
- Syntax errors stop the code before it starts running.
- Runtime errors stop the code while it is running.
- Logical errors do not crash, but the result is wrong.

## 1.3 Variables

A **variable** is a name that refers to a value.

Variables are useful because they let us remember values and reuse them in expressions.

They become especially useful when we combine them with **operators**. Operators are symbols or words that do work, like `+`, `%`, `==`, and `and`.

Try this first. Predict each line before running the cell:

```python
x = 5
print(x)
print(x + 1)
print(x)
```

The first line assigns the value `5` to the variable `x`.

After that, whenever Python sees `x`, it uses the current value of `x`.

**Important:** `x + 1` uses `x` to make a new value, but it does not change `x`. The last line still prints `5`.

### Variables and Arithmetic Operators

Arithmetic operators perform calculations. We can use them with numbers directly, or with variables that refer to numbers.

Run this code:

```python
x = 6
y = 2

print("x + y =", x + y)
print("x - y =", x - y)
print("x * y =", x * y)
print("x / y =", x / y)
```

Here are some key points:
- `+` is the addition operator.
- `-` is the subtraction operator.
- `*` is the multiplication operator.
- `/` is the division operator.
- `/` always returns a `float`, so `6 / 2` gives `3.0`, not `3`.

Python has a few more arithmetic operators that are useful with variables:

```python
base = 2
power = 3

cards = 23
players = 7

print("base ** power =", base ** power)
print("cards // players =", cards // players)
print("cards % players =", cards % players)
```

Here is what those operators mean:
- `**` is the exponentiation operator, so `2 ** 3` means `2` to the power of `3`, which is `8`.
- `//` is floor division. For positive numbers, it gives the whole-number part of division. Since `23 / 7` is `3.285...`, `23 // 7` is `3`.
- `%` is the remainder operator, also called modulo. Since `23` divided by `7` has a remainder of `2`, `23 % 7` is `2`.

A useful way to remember `//` and `%` is this: `//` tells us how many full groups we can make, and `%` tells us what is left over.

One common use of `%` is checking whether a number is even or odd. Even numbers have a remainder of `0` when divided by `2`.

```python
number = 23

print("number:", number)
print("number % 2:", number % 2)
print("Is number even?", number % 2 == 0)
```

Because `23 % 2` is `1`, the expression `number % 2 == 0` is `False`.

If a number has remainder `0` when divided by `2`, it is even. If it has remainder `1`, it is odd.

### Comparison Operators

Comparison operators compare two values and return a Boolean result: either `True` or `False`.

| Operator | Meaning |
| --- | --- |
| `==` | equal to |
| `!=` | not equal to |
| `<` | less than |
| `<=` | less than or equal to |
| `>` | greater than |
| `>=` | greater than or equal to |

```python
x = 5
y = 7

print(x == y)
print(x != y)
print(x < y)
print(x >= y)
```

To be clear:
- A single equals sign, `=`, assigns a value to a variable.
- A double equals sign, `==`, tests whether two values are equal.

So `x = 5` changes `x`, but `x == 5` asks a question and gives us `True` or `False`.

### Logical Operators

Logical operators work with Boolean values. Python has three logical operators:

| Operator | Meaning |
| --- | --- |
| `and` | `x and y` is `True` if both `x` and `y` are `True` |
| `or` | `x or y` is `True` if at least one of them is `True` |
| `not` | `not x` flips `True` to `False`, or `False` to `True` |

```python
print(True and True)
print(True and False)
print(False or True)
print(not False)
```

Logical operators are especially useful when the Boolean values come from comparisons:

```python
age = 12
has_ticket = True

print(age >= 12)
print(has_ticket)
print(age >= 12 and has_ticket)
print(age < 12 or has_ticket)
print(not has_ticket)
```

`age >= 12 and has_ticket` is `True` because both parts are `True`.

`age < 12 or has_ticket` is also `True` because at least one part is `True`.

### Assignment Operators

An assignment operator assigns a value to a variable. We already know `=`, but Python has shortcut assignment operators too.

For example, `x += y` means basically the same thing as `x = x + y`.

```python
x = 5
x = x + 3
print("x:", x)

y = 5
y += 3
print("y:", y)
```

The first three lines and the last three lines do the same kind of update.

Shortcut assignment operators are common, so it is important to be comfortable reading them.

```python
b = 7
b /= 2
print("b:", b)

c = 7
c //= 2
print("c:", c)

d = 7
d %= 2
print("d:", d)

e = 7
e **= 2
print("e:", e)
```

These are shortcuts:
- `b /= 2` means `b = b / 2`.
- `c //= 2` means `c = c // 2`.
- `d %= 2` means `d = d % 2`.
- `e **= 2` means `e = e ** 2`.

One more important detail: a variable stores the value from the moment the assignment runs. It does not remember the formula forever.

```python
apples = 4
oranges = 6
fruit = apples + oranges

print("fruit:", fruit)

apples = 10
print("fruit after changing apples:", fruit)
print("apples + oranges now:", apples + oranges)
```

`fruit` stays `10` because `fruit = apples + oranges` already ran earlier.

If we want `fruit` to use the new value of `apples`, we need to assign `fruit` again.

```python
apples = 10
oranges = 6
fruit = apples + oranges
print("new fruit:", fruit)
```
