# Lecture 01: Functions And Tests

Today we are learning how to put code inside a function and how to check that the function works.

A function is a reusable piece of code. It can take inputs and return an output.

```python
def double(number):
    return number * 2

double(5)
```

`assert` is a simple way to check if something is true. If the check is true, Python stays quiet. If it is false, Python shows an error.

```python
assert double(5) == 10
assert double(0) == 0
assert double(-3) == -6

print("All checks passed")
```
